```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Read Image and add a extra pixel border

```python
image = cv2.imread('./perito.jpeg', 0)
biggerImage = np.zeros((image.shape[0]+2, image.shape[1]+2))
plt.figure()
plt.imshow(image, cmap='gray')
plt.show()

for i in range(1, image.shape[0]-1):
    for j in range(1, image.shape[1]-1):
        biggerImage[i][j] = image[i-1][j-1]
```

Test Filters

```python
sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

blur = np.array([
    [0.0625, 0.125, 0.0625],
    [0.125,  0.25,  0.125],
    [0.0625, 0.125, 0.0625]
])

outline = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

prewitt = np.array([
    [-1, 0, 1],
    [-1,  0, 1],
    [-1, 0, 1]
])

prewitt2 = np.array([
    [-1, -1, -1],
    [0,  0, 0],
    [1, 1, 1]
])

sobel = np.array([
    [-1, 0, 1],
    [-2,  0, 2],
    [-1, 0, 1]
])

sobel2 = np.array([
    [-1, -2, -1],
    [0,  0, 0],
    [1, 2, 1]
])
```

Calculate Target Size

```python
def targetSize(N, kernel_N):
    pixels = 0

    for i in range(N):
        aux = i + kernel_N
        if aux <= N: pixels += 1
            
    return pixels
```

Convolve Function based on formula

$$F\circ I(x,y) = \sum_{j=-N}^N \sum_{i=-N}^N F(i,j) I(x+1, y+1)$$

```python
def convolve(img, kernel):
    k = kernel.shape[0]
    xSize = targetSize(img.shape[0], k)
    ySize = targetSize(img.shape[1], k)

    convolved = np.zeros(shape=(xSize, ySize))
    
    for i in range(xSize):
        for j in range(ySize):
            mat = img[i:i+k, j:j+k]
            convolved[i, j] = np.sum(np.multiply(mat, kernel))
            
    return convolved
```

```python
vertical2 = convolve(biggerImage, sobel)
hori = convolve(biggerImage, prewitt2)
vertical1 = convolve(biggerImage, prewitt)
hori2= convolve(biggerImage, sobel2)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.imshow(hori, cmap='gray')
plt.title('')
plt.subplot(1,4,2)
plt.imshow(hori2, cmap='gray')
plt.title('')
plt.subplot(1,4,3)
plt.imshow(vertical1, cmap='gray')
plt.title('')
plt.subplot(1,4,4)
plt.title('')
plt.imshow(vertical2, cmap='gray')
plt.show()
```

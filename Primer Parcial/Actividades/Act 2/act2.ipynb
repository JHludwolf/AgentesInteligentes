```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import pandas as pd
from skimage import exposure
```

```python
image = cv2.imread('dino.jpeg')
print(image.shape)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.imshow(image)
```

```python
redPixels = image[:,:,0]
print(redPixels)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(image,cmap='gray')

plt.subplot(2,2,2)
plt.title('Red Intensity Hist')
plt.hist(image[:,:,0].ravel(),256,[0,255], color="#FF0000")
plt.ylim(0, 10000)

plt.subplot(2,2,3)
plt.title('Green Intensity Hist')
plt.hist(image[:,:,1].ravel(),256,[0,255], color="#00FF00")
plt.ylim(0, 10000)

plt.subplot(2,2,4)
plt.title('Blue Intensity Hist')
plt.hist(image[:,:,2].ravel(),256,[0,255], color="#0000FF")
plt.ylim(0, 10000)

plt.show()
```

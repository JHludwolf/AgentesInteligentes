```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import skimage.io
import skimage.viewer

from PIL import Image
from scipy import ndimage as ndi
```

```python
imageGirl = cv2.imread('Girl Face.png', 0)
imageBride = cv2.imread('boda.jpeg', 0)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imageGirl, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(imageBride, cmap='gray')
```

```python
mat_identity = np.array([[1,0,0],[0,1,0],[0,0,1]])
image1 = ndi.affine_transform(imageGirl, mat_identity)
plt.figure()
plt.imshow(image1,cmap="gray")
plt.show()
```

```python
s_x, s_y = 1.5, .5
mat_scale = np.array([[s_x,0,0],[0,s_y,0],[0,0,1]])
image2 = ndi.affine_transform(imageGirl, mat_scale)
plt.figure()
plt.imshow(image2,cmap="gray")
plt.show()
```

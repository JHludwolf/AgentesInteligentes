```python
image = cv2.imread("./panamericana1.PNG")
image.shape
w, h, l = image.shape
```

```python
def scaleImage(image, s_x, s_y):
    w, h, l = image.shape
    mat_scale = np.array(
        [[s_x, 0, w/4],
        [0, s_y, h/4],
        [0, 0, 1]])

    scale = ndi.affine_transform(image[:,:,1], mat_scale)
    return scale
```

```python
def rotateImage(image, angle):
    
    theta = np.pi / (180/angle)

    mat_rotate = np.array(
        [[1, 0, w/2],
        [0, 1, h/2],
        [0, 0, 1]]) @ np.array(
            [[np.cos(theta), -np.sin(theta), 0],
            [np.sin(theta), np.cos(theta), 0],
            [0, 0, 1]]) @ np.array(
                [[1, 0, -w/2],
                [0, 1, -h/2],
                [0, 0, 1]])

    rotate = ndi.affine_transform(image[:,:,1], mat_rotate)
    return rotate
```

```python
def translateImage(image, x, y):

    mat_reflect = np.array(
        [[1, 0, w/(w/x)],
        [0, 1, h/(h/y)],
        [0, 0, 1]])
        #  @ np.array( [[1,0,w/2],[0,1,-h],[0,0,1]])

    image = cv2.warpAffine(image[:,:,1], mat_reflect, borderValue=255) # offset=(0,h)
    return image
```

```python
a = rotateImage(image, -29)
b = a < 10
c = np.logical_(a,b)

plt.figure()
plt.imshow(c, cmap='gray')
plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.imshow(image[:,:,1], cmap='gray')
plt.subplot(2,2,2)
plt.imshow(scaleImage(image, .3, .5), cmap='gray')
plt.subplot(2,2,3)
plt.imshow(rotateImage(image, -45), cmap='gray')
plt.subplot(2,2,4)
plt.imshow(translateImage(image, -500, -500), cmap='gray')
plt.show()
```
